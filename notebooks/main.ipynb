## Business Problem Understanding

**Latar Belakang**

Daegu adalah sebuah kota metropolitan di Korea Selatan, dengan status administratif sebagai “*metropolitan city*” (setara provinsi) yang langsung berada di bawah pemerintah pusat dengan populasi sebanyak 2,2 juta jiwa (per Maret 2025) dan menduduki [peringkat ke-4](https://worldpopulationreview.com/cities/south-korea) kota dengan kepadatan penduduk tertinggi di Korea Selatan, setelah Seoul, Busan, dan Incheon. 

Dalam konteks properti, fakta bahwa Daegu adalah kota besar dengan populasi jutaan dan area urban yang berkembang menunjukkan bahwa pasar apartemen (*residential apartments*) di sana memiliki kompleksitas dan potensi – namun juga tantangan seperti perbedaan kualitas lokasi, tahun pembangunan, fasilitas, dan aksesibilitas yang dapat mempengaruhi harga.

Dengan memahami konteks ini, analisis harga apartemen di Daegu menjadi penting bukan hanya untuk akademik tetapi juga aplikasi bisnis — misalnya bagi pengembang properti, agen real estate, investor atau pembeli lokal/asing yang tertarik.

Harga apartemen di kota Daegu, Korea Selatan, dipengaruhi oleh banyak faktor seperti lokasi, fasilitas sekitar, tahun pembangunan, dan ukuran apartemen. Bagi developer properti, agen real estate, atau calon pembeli, penting untuk memahami faktor-faktor apa saja yang berpengaruh signifikan terhadap harga jual apartemen agar dapat membuat keputusan yang lebih akurat.

Namun, harga apartemen di Daegu sangat bervariasi antar area, bangunan, dan fasilitas.
Bagi agen real estate (pada kasus ini sebagai **Stakeholder**), menentukan harga jual atau nilai pasar yang tepat untuk setiap unit apartemen menjadi tantangan besar — kesalahan harga dapat berakibat pada:

Selain itu, agen sering harus mengandalkan intuisi dan pengalaman pribadi, bukan data objektif, dalam menentukan estimasi harga. Hal ini menyebabkan keputusan bisnis menjadi tidak konsisten dan sulit diukur secara kuantitatif.

**Metric Evaluation**

Evaluasi metrik yang akan digunakan adalah RMSE, MAE, dan MAPE, di mana RMSE adalah nilai rataan akar kuadrat dari error, MAE adalah rataan nilai absolut dari error, sedangkan MAPE adalah rataan persentase error yang dihasilkan oleh model regresi. Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, berarti model semakin akurat dalam memprediksi harga sewa sesuai dengan limitasi fitur yang digunakan. 

Selain itu, kita juga bisa menggunakan nilai R-squared atau adj. R-squared jika model yang nanti terpilih sebagai final model adalah model linear. Nilai R-squared digunakan untuk mengetahui seberapa baik model dapat merepresentasikan varians keseluruhan data. Semakin mendekati 1, maka semakin fit pula modelnya terhadap data observasi. Namun, metrik ini tidak valid untuk model non-linear.

Dalam konteks bisnis: jika MAE misalnya ±₩40000 (won) atau ±Rp 450 juta, maka artinya rata-rata prediksi meleset sebanyak itu — kita harus menilai apakah itu masih “cukup akurat” untuk tujuan bisnis (apakah decision-making masih bisa diterima).

#### 4. *Korelasi Fitur dan Target*

```python
# Korelasi fitur Numerik

corr_spearman = df_cleaned.corr(method='spearman', numeric_only=True)['SalePrice']
result = pd.DataFrame(corr_spearman)
result = result.drop('SalePrice', axis=0)
result = result.sort_values(by='SalePrice', ascending=False)

display(result)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(result, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Spearman Fitur terhadap Target (SalePrice)', fontsize=13)
plt.xlabel('Metode Korelasi')
plt.ylabel('Fitur')
plt.show()
```

Faktor paling berpengaruh positif terhadap harga apartemen adalah ukuran unit (`Size(sqf)`) — korelasi sangat kuat (+0.65). Ini berarti peningkatan ukuran akan secara konsisten menaikkan harga jual.

Fasilitas internal (`N_FacilitiesInApt`) dan tahun pembangunan (`YearBuilt`) juga berperan besar, menunjukkan bahwa modernity dan lifestyle quality menjadi nilai tambah utama bagi pembeli.

Aksesibilitas parkir (`N_Parkinglot(Basement)`) menambah nilai jual, tetapi tidak sekuat faktor-faktor sebelumnya.

Fitur lingkungan sekitar tertentu justru berkorelasi negatif dengan harga — artinya, tidak semua lokasi ramai adalah lokasi yang bernilai tinggi. Apartemen di area yang terlalu komersial atau dekat kampus/universitas mungkin dihargai lebih rendah karena target pasarnya berbeda.

```python
# Korelasi fitur kategorik

categorical_columns = ['HallwayType', 'TimeToSubway', 'SubwayStation']
kruskal_results = []

for col in categorical_columns:
    # Group SalePrice by each category
    groups = [group['SalePrice'].values for name, group in df_cleaned.groupby(col)]
    
    # Perform Kruskal–Wallis test
    stat, p_value = kruskal(*groups)
    
    kruskal_results.append({
        'Feature': col,
        'Kruskal_Statistic': stat,
        'p-value': p_value
    })
    kruskal_df = pd.DataFrame(kruskal_results).sort_values(by='p-value')
display(kruskal_df)
for i, row in kruskal_df.iterrows():
    feature = row['Feature']
    p = row['p-value']
    if p < 0.05:
        print(f"- {feature}: p-value = {p:.4f} → terdapat perbedaan signifikan harga antar kategori.")
    else:
        print(f"- {feature}: p-value = {p:.4f} → tidak ada bukti kuat bahwa kategori memengaruhi harga.")
```

Semua fitur kategorikal memiliki p-value < 0.05,
artinya terdapat perbedaan signifikan pada harga jual antar kategori dalam setiap variabel.

#### **Feature Engineering Setup**

```python
# Create Transformation Pipeline

# Numerical Features Pipeline
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler()) # Sudah dicoba StandardScaler, hasil lebih bagus pada Robust
])

# Categorical Features Pipeline (OHE)
ohe_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore", drop="first"))
])

display(num_pipeline, ohe_pipeline)
```

```python
# Create Feature Transformer
num_features = ['N_FacilitiesNearBy(PublicOffice)', 'N_SchoolNearBy(University)', 'YearBuilt','N_FacilitiesInApt','Size(sqf)']
ohe_features = ['HallwayType', 'TimeToSubway', 'SubwayStation']

# Baseline Preprocessor
baseline_preprocessor = ColumnTransformer([
    ("num_pipeline", num_pipeline, num_features),
    ("ohe_pipeline", ohe_pipeline, ohe_features),
], remainder="drop", verbose_feature_names_out=False)

display(baseline_preprocessor)
```

```python
X_train_pre = baseline_preprocessor.fit_transform(X_train)
fe_overview = pd.DataFrame(X_train_pre, columns=baseline_preprocessor.get_feature_names_out())
```

```python
print("Before Preprocessing:")
display(X_train.head(1))

print("After Preprocessing:")
display(fe_overview.head(1))
```

#### **Choose a Benchmark Model**

```python
# Individual Models (Baseline)
regressors = {
    "Linear Regression": LinearRegression(),
    "KNN Regressor": KNeighborsRegressor(),
    "Decision Tree Regressor": DecisionTreeRegressor(random_state=42),
    "Random Forest Regressor": RandomForestRegressor(random_state=42, n_jobs=-1),
    "AdaBoost Regressor": AdaBoostRegressor(random_state=42),
    "XGBoost Regressor": XGBRegressor(random_state=42, n_jobs=4, verbosity=0)
}
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
def neg_to_pos(arr):
    # cross_val_score returns negative for loss metrics; ubah jadi positive
    return -np.array(arr)

results = []

for name, model in regressors.items():
    # buat pipeline lengkap: preprocessor (ColumnTransformer) + model
    pipe = Pipeline([
        ("preprocessor", baseline_preprocessor),
        ("model", model)
    ])
    
    # RMSE (neg_root_mean_squared_error)
    scores_rmse = cross_val_score(pipe, features, target, cv=cv,
                                  scoring='neg_root_mean_squared_error', n_jobs=-1, error_score='raise')
    scores_rmse = -scores_rmse  # ubah ke positive RMSE
    
    # MAE
    scores_mae = cross_val_score(pipe, features, target, cv=cv,
                                 scoring='neg_mean_absolute_error', n_jobs=-1, error_score='raise')
    scores_mae = -scores_mae
    
    # MAPE
    scores_mape = cross_val_score(pipe, features, target, cv=cv,
                                  scoring='neg_mean_absolute_percentage_error', n_jobs=-1, error_score='raise')
    scores_mape = -scores_mape  # values seperti 0.17 -> 17%
    
    results.append({
        "Model": name,
        "Mean_RMSE": scores_rmse.mean(),
        "Std_RMSE": scores_rmse.std(),
        "Mean_MAE": scores_mae.mean(),
        "Std_MAE": scores_mae.std(),
        "Mean_MAPE": scores_mape.mean(),
        "Std_MAPE": scores_mape.std(),
        "RMSE_CV_Array": scores_rmse,  # optional, untuk inspeksi
    })

results_df = pd.DataFrame(results).sort_values(by="Mean_RMSE")
# konversi MAPE ke persen (opsional) agar lebih mudah dibaca
results_df["Mean_MAPE_pct"] = results_df["Mean_MAPE"] * 100
results_df["Std_MAPE_pct"] = results_df["Std_MAPE"] * 100


display(results_df[['Model','Mean_RMSE','Std_RMSE','Mean_MAE','Std_MAE','Mean_MAPE_pct','Std_MAPE_pct']])
```

Berdasarkan data diatas, setelah dilakukan cross-validation sebanyak 5 kali, dapat dilihat nilai RMSE terendah pada XGBoost dengan nilai 46177 won (19% dari mean), nilai MAE terendah pada RandomForest dengan nilai 37291 won (14% dari mean), dan nilai MAPE terendah juga pada RandomForest dengan nilai 19,04%.

Terdapat perbedaan cukup signifikan antara nilai RMSE dan MAE, di mana nilai RMSE lebih tinggi karena nilai residuals atau error-nya dikuadratkan terlebih dahulu sebelum dirata-ratakan. Hal ini menyebabkan RMSE akan memberikan 'weight' yang lebih tinggi untuk nilai error yang besar. Dengan kata lain, terdapat nilai-nilai error yang besar yang dihasilkan oleh semua algoritma yang digunakan, sehingga ada perbedaan signifikan antara nilai RMSE dan MAE.

- Berdasarkan nilai RMSE, MAE dan MAPE: XGBoost adalah model terbaik. Model terbaik kedua adalah RandomForest.
- Selanjutnya, akan dilakukan prediksi pada test set dengan 2 benchmark model terbaik, yaitu RandomForest dan XGBoost. 

#### **Predict to Test Set with the Benchmark Model**

Prediksi pada test set dengan menggunakan model XGBoost dan Random Forest

```python
rf_model = RandomForestRegressor(random_state=42, n_jobs=-1)
xgb_model = XGBRegressor(random_state=42, n_jobs=4, verbosity=0)
rf_pipe = Pipeline([
    ('preprocessor', baseline_preprocessor),
    ('model', rf_model)
])

xgb_pipe = Pipeline([
    ('preprocessor', baseline_preprocessor),
    ('model', xgb_model)
])
rf_pipe.fit(X_train, y_train)
xgb_pipe.fit(X_train, y_train)

y_pred_rf = rf_pipe.predict(X_test)
y_pred_xgb = xgb_pipe.predict(X_test)

# --- Combine predictions into one DataFrame ---
compare_df = pd.DataFrame({
    "Actual": y_test.values,
    "Pred_RF": y_pred_rf,
    "Pred_XGB": y_pred_xgb
})

# --- Compute error for each model ---
compare_df["Error_RF"] = compare_df["Actual"] - compare_df["Pred_RF"]
compare_df["Error_XGB"] = compare_df["Actual"] - compare_df["Pred_XGB"]
compare_df["AbsError_RF"] = np.abs(compare_df["Error_RF"])
compare_df["AbsError_XGB"] = np.abs(compare_df["Error_XGB"])

# --- Metrics summary ---
def model_metrics(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return rmse, mae, mape

rf_rmse, rf_mae, rf_mape = model_metrics(compare_df["Actual"], compare_df["Pred_RF"])
xgb_rmse, xgb_mae, xgb_mape = model_metrics(compare_df["Actual"], compare_df["Pred_XGB"])

metrics_summary = pd.DataFrame({
    "Model": ["Random Forest", "XGBoost"],
    "RMSE": [rf_rmse, xgb_rmse],
    "MAE": [rf_mae, xgb_mae],
    "MAPE (%)": [rf_mape, xgb_mape]
})

print("Test Set Performance Comparison")
display(metrics_summary.round(2))

plt.figure(figsize=(10,5))
sns.kdeplot(compare_df["Actual"], label="Actual", color="black", lw=2)
sns.kdeplot(compare_df["Pred_RF"], label="Random Forest", color="blue", lw=2)
sns.kdeplot(compare_df["Pred_XGB"], label="XGBoost", color="orange", lw=2)
plt.title("Actual vs Predicted Price Distributions (Test Set)", fontsize=13)
plt.xlabel("Sale Price (won)")
plt.legend()
plt.show()
```

- Kedua model sudah dapat memprediksi harga apartment dengan cukup baik, dimana grafik prediksi sudah mengikuti pola dari aktual test, kedua model pun memiliki pola yang identik.

#### **Hyperparameter Tuning**

Karena model XGBoost memiliki performa yang sedikit lebih baik dari RandomForest, maka selanjutnya kita akan menggunakan model XGBoost ini sebagai model akhir. Pada tahap ini, kita akan melakukan hyperparameter tuning pada model XGBoost dengan harapan dapat meningkatkan performa model.

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)
scoring = 'neg_root_mean_squared_error'
```

```python
# RandomForest Tuning
rf_base = RandomForestRegressor(random_state=42, n_jobs=-1)
# rf_ttr = TransformedTargetRegressor(regressor=rf_base, func=np.log1p, inverse_func=np.expm1)
pipe_rf = Pipeline([
    ('preprocessor', baseline_preprocessor),
    ('model', rf_base)
])

rf_param_dist = {
    # note: when using TTR, regressor params are under 'model__regressor__<param>'
    'model__n_estimators': [100, 200, 400],
    'model__max_depth': [None, 6, 10, 20],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
    'model__max_features': ['sqrt', 'auto', 0.5]
}

rs_rf = RandomizedSearchCV(
    estimator=pipe_rf,
    param_distributions=rf_param_dist,
    n_iter=100,
    scoring=scoring,
    cv=cv,
    random_state=42,
    n_jobs=-1,
    verbose=2,
    return_train_score=False
)
rs_rf.fit(X_train, y_train)
best_rf_pipe = rs_rf.best_estimator_
best_params = rs_rf.best_params_

mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
mape_scorer = make_scorer(mape, greater_is_better=False)
scores_rmse_best = cross_val_score(best_rf_pipe, features, target, cv=cv, scoring='neg_root_mean_squared_error', n_jobs=-1)
scores_mae_best = cross_val_score(best_rf_pipe, features, target, cv=cv, scoring='neg_mean_absolute_error', n_jobs=-1)
scores_mape_best = cross_val_score(best_rf_pipe, features, target, cv=cv, scoring=mape_scorer, n_jobs=-1)

tuned_results = {
    "Model": "RF Tuned",
    "Mean_RMSE": -scores_rmse_best.mean(),
    "Std_RMSE": scores_rmse_best.std(),
    "Mean_MAE": -scores_mae_best.mean(),
    "Std_MAE": scores_mae_best.std(),
    "Mean_MAPE%": -scores_mape_best.mean(),
    "Std_MAPE%": scores_mape_best.std()
}

comparison = pd.DataFrame([tuned_results])
display(comparison.round(2))
print(best_params)
```

```python
# Tuning XGBoost
xgb_base = XGBRegressor(random_state=42, objective='reg:squarederror', n_jobs=4, verbosity=0)

xgb_ttr = TransformedTargetRegressor(regressor=xgb_base, func=np.log1p, inverse_func=np.expm1)

pipe_xgb = Pipeline([
    ('preprocessor', baseline_preprocessor),
    ('model', xgb_ttr)
])

xgb_param_dist = {
    'model__regressor__n_estimators': [100, 200, 150, 180],
    'model__regressor__learning_rate': [0.3, 0.07, 0.05, 0.04],
    'model__regressor__max_depth': [6, 4, 5, 7],
    'model__regressor__subsample': [1.0, 0.8, 0.7],
    'model__regressor__colsample_bytree': [1, 0.7, 0.8, 0.6],
    # optional regularization
    'model__regressor__reg_alpha': [0, 0.01, 0.1],
    'model__regressor__reg_lambda': [1, 2, 2.25]
}

rs_xgb = RandomizedSearchCV(
    estimator=pipe_xgb,
    param_distributions=xgb_param_dist,
    n_iter=100,
    scoring=scoring,
    cv=cv,
    random_state=42,
    n_jobs=-1,
    verbose=2,
    return_train_score=False
)

rs_xgb.fit(X_train, y_train)
best_xgb_pipe = rs_xgb.best_estimator_
best_params = rs_xgb.best_params_

mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
mape_scorer = make_scorer(mape, greater_is_better=False)
scores_rmse_best = cross_val_score(best_xgb_pipe, features, target, cv=cv, scoring='neg_root_mean_squared_error', n_jobs=-1)
scores_mae_best = cross_val_score(best_xgb_pipe, features, target, cv=cv, scoring='neg_mean_absolute_error', n_jobs=-1)
scores_mape_best = cross_val_score(best_xgb_pipe, features, target, cv=cv, scoring=mape_scorer, n_jobs=-1)

tuned_results = {
    "Model": "XGBoost Tuned",
    "Mean_RMSE": -scores_rmse_best.mean(),
    "Std_RMSE": scores_rmse_best.std(),
    "Mean_MAE": -scores_mae_best.mean(),
    "Std_MAE": scores_mae_best.std(),
    "Mean_MAPE%": -scores_mape_best.mean(),
    "Std_MAPE%": scores_mape_best.std()
}

comparison = pd.DataFrame([tuned_results])
display(comparison.round(2))
print(best_params)
```

#### **Predict to Test Set with the Tuned Model**

Melakukan prediksi pada test set dengan menggunakan hyperparameter terpilih.

```python
# Predict on test data
y_pred_rf_tuned = best_rf_pipe.predict(X_test)
y_pred_xgb_tuned = best_xgb_pipe.predict(X_test)

# Evaluate metrics
def evaluate_model(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return rmse, mae, mape

rf_rmse, rf_mae, rf_mape = evaluate_model(y_test, y_pred_rf_tuned)
xgb_rmse, xgb_mae, xgb_mape = evaluate_model(y_test, y_pred_xgb_tuned)

# Print updated performance metrics
updated_metrics = pd.DataFrame({
    "Model": ["Random Forest (Tuned)", "XGBoost (Tuned)"],
    "RMSE": [rf_rmse, xgb_rmse],
    "MAE": [rf_mae, xgb_mae],
    "MAPE (%)": [rf_mape, xgb_mape]
})

# --- Display results ---
print("Tuned Model Test Metrics")
display(updated_metrics.round(2))

plt.figure(figsize=(10,5))
sns.kdeplot(y_test, label="Actual", color="black", lw=2)
sns.kdeplot(y_pred_rf_tuned, label="Random Forest (Tuned)", color="blue", lw=2)
sns.kdeplot(y_pred_xgb_tuned, label="XGBoost (Tuned)", color="orange", lw=2)

plt.title("Actual vs Predicted Price Distributions (Test Set, After Tuning)", fontsize=13, fontweight="bold")
plt.xlabel("Sale Price (won)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

**Model Trustworthiness**

- Model dapat dipercaya & digunakan ketika:
    - Data properti baru masih berada dalam karakteristik yang sama seperti data Daegu (misal: ukuran, fasilitas, tipe hallway, jarak ke subway).
    - Agen ingin melakukan estimasi harga jual awal untuk properti baru yang belum memiliki nilai pasar
    - Digunakan sebagai bantuan keputusan (decision support tool), bukan penentu tunggal harga
    - Tujuannya adalah estimasi kisaran harga (range), bukan angka pasti.
- Model kurang dapat dipercaya ketika:
    - Data baru berbeda secara signifikan dari data pelatihan (misalnya apartemen di kota lain, gedung super premium, atau kondisi ekonomi berubah drastis).
    - Terdapat fitur baru yang belum pernah muncul di data pelatihan (misal fitur “ocean view”, “smart home”, dsb).
    - Harga pasar mengalami anomali makroekonomi, seperti perubahan kebijakan properti, suku bunga, atau pandemi.

## Recommendation

1. Strategi Penentuan Harga:
    - Gunakan model XGBoost yang sudah dituning sebagai alat prediksi harga dasar.
    - Fokuskan penilaian harga berdasarkan ukuran unit, tahun bangun, dan tipe hallway.
    - Untuk apartemen dengan tipe hallway terraced, berikan premium price karena segmen ini memiliki willingness-to-pay lebih tinggi.

2. Strategi Pemasaran Properti
    - Utamakan fitur interior & fasilitas (gym, taman, balkon, parkir) dalam materi promosi, bukan hanya lokasi.
    - Untuk unit dengan hallway terraced atau mixed, tampilkan desain hallway dan privasi sebagai nilai jual utama.- Gunakan insight ini untuk mengarahkan calon pembeli ke apartemen dengan fitur bernilai tinggi, bukan hanya lokasi populer.

3. Rekomendasi untuk improvement model/project
    - Perluasan Dataset dengan menambahkan data dari wilayah lain (misal: Busan, Seoul, Incheon) untuk membuat model yang lebih general.
    - Integrasi Faktor Eksternal dengan menambahkan variabel seperti lokasi geospasial, jarak ke pusat bisnis, tingkat inflasi, trend pasar properti tahunan.
    - Implementasi Dashboard Prediksi dengan membuat aplikasi web sederhana (misalnya Streamlit/Dash) agar agen real estate bisa:
        - Input fitur apartemen baru,
        - Mendapatkan prediksi harga instan,
        - Melihat analisis faktor yang paling berpengaruh.
    - Lakukan retraining setiap 6–12 bulan dengan data terbaru untuk menyesuaikan dinamika harga pasar.
